```python
file_path = 'Dataset Inlämningsuppgift.xlsx'
```

```python
data = pd.read_excel(file_path)
```

```python
log_test = np.log(data['Profit'])
```

```python
# dataframe with top 5 profit
top_5_profit = data.sort_values(by='Profit', ascending=False).head(5)
```

```python
top_5_profit
```

```python
# show the top 5 profit in a chart
plt.figure(figsize=(12, 6))
plt.bar(top_5_profit['Product Name'], top_5_profit['Profit'], color='blue', alpha=0.75)
plt.xlabel('Product Name')
# x axis needs to be rotated to be able to read the product names
plt.xticks(rotation=45, ha='right')
plt.ylabel('Profit')
plt.title('Top 5 Profit')
plt.show()
```
